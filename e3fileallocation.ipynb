```
dir=['']*50
rear=0


dir_name=input("Enter the directory name: ")
while(1):
    count=0
    cd=0
    print("Enter your choice: \n 1.Add a file \n 2.Delete a file \n 3.Search file \n 4.Display files \n 5.Exit\n")
    ch=int(input())
    if(ch==1):
        flag=1
        fname=input("Enter filename: ")
        for i in range (rear):
            if(dir[i]==fname):
                print("Similar filename already exists")
                flag=0
                break
        if(flag!=0):
            dir[rear]=fname
            rear+=1
        

    elif(ch==2):
        dname=input("\n Enter filename:")
        for i in range(rear):
            if(dir[i]==dname):
                dir[i]=''
                cd+=1
        if(cd==0):
            print("File Not Found!\n")
        

    elif(ch==3):
        x=0
        dname=input("\n Enter filename:")
        for i in range(rear):
            if(dir[i]==dname):
                print("File found at position",i)
            else:
                x=x+1
        if(x==rear):
            print("File not found")
        
       
    elif(ch==4):
        print("")
        print("The files in directory",dir_name,end='')
        print(" are listed below:")
        for i in range(rear):
            if(dir[i]!=''):
                print("-->",end='')
                print(dir[i])
                count+=1
        if(count==0):
            print(" No Files Found ")
        print("")
    
        
    
    else:
        exit();
```

```
main_directory=[]
main=[0]*50
x=0
index=0
while(1):
    print("Choose your Option: \n 1.Add sub-directory\n 2.Delete sub-directory\n 3.Add files to a sub-directory \n 4.Delete files from the sub-directory\n5.Display\n 6.Exit\n")
    ch=int(input())
    if(ch==1):
        print("Enter the number of sub-directories")
        n=int(input())
        for i in range(n):
            print("Enter the sub-directory name:")
            s_name=input()
            if s_name not in main:
                main[i+x]=s_name
                s_name=[]
                main_directory.append(s_name)
            else:
                print(s_name,"sub-directory already exists")
        x=x+n        

    elif(ch==2):
        print("Enter the sub-directory to be deleted")
        s_name=input()
        if s_name in main:
            index=main.index(s_name)
            main.pop(index)
            main_directory.pop(index)
            print("Deleted the sub-directory",s_name)
        else:
            print("No such sub-directory exists")
        #for i in range(len(main_directory)):
            #print(main_directory[i])
    elif(ch==3):
        print("Enter the sub-directory to which file has to be added")
        s_name=input()
        if s_name in main:
            index=main.index(s_name)
            print("Enter the number of files to be added")
            m=int(input())
            for j in range(m):
                print("Enter the file name")
                f_name=input()
                if f_name not in main_directory[index]:
                    main_directory[index].append(f_name)
                else:
                    print("File",f_name,"already exists")
        else:
            print("No such sub-directory")

    elif(ch==4):
        print("Enter the sub-directory from which file has to be deleted")
        s_name=input()
        if s_name in main:
            index=main.index(s_name)
            print("Enter the file name")
            f_name=input()
            if f_name in main_directory[index]:
                main_directory[index].remove(f_name)
            else:
                print("No such file exists")
        else:
            print("No such sub-directory")
       
    elif(ch==5):
        print("Existing Sub-directories and files in it")
        for i in range(len(main_directory)):
            print(main[i],":",main_directory[i])
        
    
    else:
        exit();
```
